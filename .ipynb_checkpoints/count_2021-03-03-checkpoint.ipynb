# Read the Data

```python
AccFiles = os.listdir('./Data/allLabeled/Acceleration/')
fileNames = [i[:-4] for i in AccFiles if i[-4:]=='.csv'] 

accDict = {}
for i, key in enumerate(fileNames):
    accDict[key] = pd.read_csv('./Data/allLabeled/Acceleration/'+AccFiles[i], header=None)

trueSegDict = pd.read_excel('./Data/allLabeled/True segments/segments.xlsx', engine='openpyxl', sheet_name=None)
```

## Apply Filter

```python
SR = accDict['Saeb'].values[1,0]
n=4; fc=4; w=fc/(SR/2)

accFiltDict = {}
segs = {}
for key in accDict.keys():
    b, a = signal.butter(n, w, 'low')
    accFiltDict[key] = signal.filtfilt(b, a, x=accDict[key].values[2:,:], axis=0)
    
    segs[key] = trueSegDict[key].values[:,1].astype('float64')*SR

accFiltDict['Saeb'] = accFiltDict['Saeb'][1210:46980,:]
tmp1 = np.sqrt(accFiltDict['Saeb'][:,0]**2 + accFiltDict['Saeb'][:,1]**2 + accFiltDict['Saeb'][:,2]**2)
accFiltDict['Saeb'] = np.hstack((accFiltDict['Saeb'], tmp1.reshape(tmp1.shape[0], 1)))

accFiltDict['Sahand'] = accFiltDict['Sahand'][5012:59980,:]
tmp2 = np.sqrt(accFiltDict['Sahand'][:,0]**2 + accFiltDict['Sahand'][:,1]**2 + accFiltDict['Sahand'][:,2]**2)
accFiltDict['Sahand'] = np.hstack((accFiltDict['Sahand'], tmp2.reshape(tmp2.shape[0], 1)))
```

## Visualize Filtered Signal and True Segments

```python
f, ax = plt.subplots(2,1, figsize=(13,7))
ax[0].plot(accFiltDict['Saeb'][:,0], c='r')
ax[0].plot(accFiltDict['Saeb'][:,1], c='g')
ax[0].plot(accFiltDict['Saeb'][:,2], c='b')
ax[0].set_title('Saeb', fontsize=20)
for v in segs['Saeb']:
    ax[0].axvline(v, c='b')

ax[1].plot(accFiltDict['Sahand'][:,0], c='r')
ax[1].plot(accFiltDict['Sahand'][:,1], c='g')
ax[1].plot(accFiltDict['Sahand'][:,2], c='b')
ax[1].set_title('Sahand', fontsize=20)
for v in segs['Sahand']:
    ax[1].axvline(v, c='b')
f.savefig('./blah.png')
```

## Search the Signal using Manual Motif and DTW Distance

```python
startTime = time.time()
distDTWLiftSaeb = []
for i in range(len(liftMagFiltSaeb)-len(liftMotif)):
    tmpDTW, _ = fastdtw(liftMotif, liftMagFiltSaeb[i:i+len(liftMotif)], dist=euclidean)
    distDTWLiftSaeb.append(tmpDTW)
    
distDTWLiftSahand = []
for i in range(len(liftMagFiltSahand)-len(liftMotif)):
    tmpDTW, _ = fastdtw(liftMotif, liftMagFiltSahand[i:i+len(liftMotif)], dist=euclidean)
    distDTWLiftSahand.append(tmpDTW)
print('Elapsed Time = ', time.time() - startTime)
```

```python
denominator = 2; percentile = 5
signal_min = np.percentile(distDTWLiftSaeb, percentile)
signal_max = np.percentile(distDTWLiftSaeb, 100-percentile)
mph = (signal_max - signal_min)/denominator
valleysLiftSaeb = detect_peaks(distDTWLiftSaeb, mph, valley=True, mpd=200)


signal_min = np.percentile(distDTWLiftSahand, percentile)
signal_max = np.percentile(distDTWLiftSahand, 100-percentile)
mph = (signal_max - signal_min)/denominator
valleysLiftSahand = detect_peaks(distDTWLiftSahand, mph, valley=True, mpd=200)



#### plot
f, ax = plt.subplots(4,1, figsize=(13,14), sharex=True)
ax[0].plot(distDTWLiftSaeb)
ax[0].scatter(valleysLiftSaeb, np.array(distDTWLiftSaeb)[valleysLiftSaeb], c='r', s=10)
ax[0].set_title('Distance and selected vallyes Saeb')

ax[1].plot(liftMagFiltSaeb)
color = np.repeat([['r', 'g', 'b']], len(valleysLiftSaeb)//3+3, axis=0).flatten()[:len(valleysLiftSaeb)]
for st, col in zip(valleysLiftSaeb, color):
    ax[1].plot(range(st, st+len(liftMotif)), liftMagFiltSaeb[st:st+len(liftMotif)], c=col, alpha=0.3, linewidth=7)



ax[2].plot(distDTWLiftSahand)
ax[2].scatter(valleysLiftSahand,
              np.array(distDTWLiftSahand)[valleysLiftSahand], c='r', s=10)
ax[2].set_title('Distance and selected vallyes Sahand')

ax[3].plot(liftMagFiltSahand)
color = np.repeat([['r', 'g', 'b']], len(valleysLiftSahand)//3+3, axis=0).flatten()[:len(valleysLiftSahand)]
for st, col in zip(valleysLiftSahand, color):
    ax[3].plot(range(st, st+len(liftMotif)), liftMagFiltSahand[st:st+len(liftMotif)], c=col, alpha=0.3, linewidth=7)
# ax[1].set_xlim([-10, 2000])

plt.show()
```
