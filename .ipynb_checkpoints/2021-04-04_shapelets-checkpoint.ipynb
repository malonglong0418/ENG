# Read Wrist and Annotation Data

## Fuctions for Reading the Data

```python
def readData(accDir, annotFile):
    files = os.listdir(accDir)
    files_csv = [f for f in files if f[-3:] == 'csv']
    empatica_dict = dict()
    for f in files_csv:
        data = np.genfromtxt(accDir+f, delimiter=',') # creates numpy array for each Empatica acc csv file
        key = int(float(f.strip("ACC.csv")))
        empatica_dict[key] = data
    tmp = pd.read_excel(annotFile, sheet_name=None)
    annot_dict = dict(zip(tmp.keys(), [i.dropna() for i in tmp.values()])) # Remove the rows with NaN values (some with ladder 2 missing)
    return empatica_dict, annot_dict

def getLabeledDict(empatica_dict, annot_dict, subject_ids):
    labeled_dict = {}; taskInd_dict = {}
    for id in subject_ids:
        start_time = int(empatica_dict[id][0,0])
        acc = empatica_dict[id][2:,:]
        label = list(map(lambda i: i.replace("_end", "").replace("_start", ""), annot_dict['P'+ str(id)].taskName.tolist()))
        task_time= list(map(lambda i: time.mktime(datetime.datetime.strptime(i[:6] + '20' + i[6:], "%m/%d/%Y %H:%M:%S").timetuple()),
                            annot_dict['P'+ str(id)].startTime_global.tolist()))
        task_ind = [int(x - start_time)*SR for x in task_time]
        taskInd_dict[id] = task_ind
        label_tmp = np.empty(acc.shape[0], dtype=object)
        for i, (j, k) in enumerate(zip(task_ind[0::2], task_ind[1::2])):
            tmpInd = 2*i
            label_tmp[j:k] = label[tmpInd]
        acc_mag = np.sqrt(np.sum(acc**2, axis=1))[:,None]
        accel = np.hstack((acc, acc_mag))
        labeled_dict[id] = pd.DataFrame(np.hstack((accel, label_tmp.reshape(label_tmp.shape[0],1))), columns=['X', 'Y', 'Z', 'Mag', 'label'])
    return labeled_dict, taskInd_dict
```

## Read the Data

```python
sepAccDict, sepAnnotDict = readData(accDir='./Data/50_subs/Acc Data/separate/', annotFile='./Data/50_subs/Annotation Data/separate.xlsx')
SR=int(sepAccDict[8][1,0])

sepSubIDs = list(range(8,45))
sepSubIDs.remove(27) # does not have lift
sepLabeledDict_, sepTaskIndDict = getLabeledDict(sepAccDict, sepAnnotDict, sepSubIDs)
```

## Apply Low Pass Filter

```python
# Apply Filter on All Subjects
n=4; fc=2; w=fc/(SR/2)
b, a = signal.butter(n, w, 'low')
sepLabeledDict_filtered = dict(map(lambda key: (key, signal.filtfilt(b, a, x=sepLabeledDict_[key].drop(columns='label'), axis=0)), sepLabeledDict_.keys()))
# back to DF and add label
sepLabeledDict_filtered_dfs = dict(map(lambda key: (
                                                        key, pd.DataFrame(sepLabeledDict_filtered[key],columns=['X', 'Y', 'Z', 'Mag']).assign(label=sepLabeledDict_[key].label)
                                                    ), sepLabeledDict_filtered.keys()))
# Remove data without label
filt_noNA_dict = dict(map(lambda key: (key, sepLabeledDict_filtered_dfs[key].dropna()), sepLabeledDict_filtered_dfs.keys()))
```

## Create Netsed Dataframe for sktime Analysis

```python
df_ = filt_noNA_dict[8]
df = df_[df_.label.isin(['hoist', 'ladder1', 'ladder2', 'lift'])]
window_len = 300

tmp1 = df.groupby('label').apply(lambda x: x.Y.values[:(x.shape[0]//window_len)*window_len].reshape(x.shape[0]//window_len, -1))
tmp2 = np.vstack(tmp1)
X = from_2d_array_to_nested(tmp2, columns=['windows'])

y_list_ = []
[y_list_.extend(elem.shape[0]*[label]) for elem, label in zip(tmp1, tmp1.index)]
y_list = [elem.replace('ladder1', 'ladder').replace('ladder2', 'ladder') for elem in y_list_]

y = np.array(y_list)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, shuffle=True, random_state=2021)

start_time = time.time()
time_contract_in_mins = 20
initial_num_shapelets_per_case = 10
verbose = False

st = ContractedShapeletTransform(
    min_shapelet_length=20,
    max_shapelet_length=150,
    time_contract_in_mins=time_contract_in_mins,
    num_candidates_to_sample_per_case=initial_num_shapelets_per_case,
    verbose=verbose,
)
st.fit(X_train, y_train)
print('elapsed time = {}'.format(time.time() - start_time))
```

## Find Most Informative Shapelet for each Label

```python
# tmp1 = np.array([y_train[s.series_id] for s in st.shapelets])
# label_dict = dict([(elem, np.min(np.where(tmp1==elem))) for elem in np.unique(tmp1)])
# shapelet_dict = dict([(key, st.shapelets[value].data[0]) for key, value in label_dict.items()])
```

```python
list_ = []
for s in st.shapelets:
    tmp = pd.DataFrame([{'shapelet': X_train.iloc[s.series_id, 0][s.start_pos : s.start_pos + s.length],
                         'label': y_train[s.series_id], 'start_pos': s.start_pos, 'length': s.length, 'info_gain': s.info_gain}])
    list_.append(tmp)
shapelet_df_ = pd.concat(list_)

shapelet_df = shapelet_df_.sort_values('info_gain', ascending=False).drop_duplicates(['label'])
shapelet_dict = dict()
for label in shapelet_df.label:
    shapelet_dict[label] = shapelet_df.shapelet[shapelet_df.label==label].values[0].values
```

```python
f, ax = plt.subplots(1,3, figsize=(12,4), sharey=True)
ax[0].plot(shapelet_dict['lift'])
ax[0].set_title('Lift,  Length:' + str(shapelet_dict['lift'].shape[0]))
ax[1].plot(shapelet_dict['hoist'])
ax[1].set_title('Hoist,  Length:' + str(shapelet_dict['hoist'].shape[0]))
ax[2].plot(shapelet_dict['ladder'])
ax[2].set_title('Ladder,  Length:' + str(shapelet_dict['ladder'].shape[0]))
f.suptitle('Most Informative Shapelets', fontsize=20)
f.tight_layout()
plt.show()
```

## Perform Similarity Search with each Shapelet

```python
startTime = time.time()
label = 'lift'
sig = df[df.label==label].Y.values
query = shapelet_dict[label]
dist = []
for i in range(len(sig)-len(query)):
    dist.append(euclidean(query, sig[i:i+len(query)]))

f, ax = plt.subplots(3,1,figsize=(20,10))
ax[0].plot(query)
ax[0].set_title(label + '  Shapelet', fontsize=20)
ax[0].set_xlim([-100, 400])
ax[1].plot(sig)
ax[1].set_title(label, fontsize=20)
ax[2].plot(dist)
ax[2].set_title('Euclidean Distance', fontsize=20)
f.tight_layout()
plt.show()
```
